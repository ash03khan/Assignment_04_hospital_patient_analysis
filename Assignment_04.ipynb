```python
import pandas as pd
import numpy as np
```

In this step, we import the required libraries for data analysis.

Pandas is used for handling structured data such as tables. It provides powerful tools for filtering, grouping, merging, and cleaning data. NumPy is used for numerical operations and supports efficient calculations on large datasets.

These libraries are standard tools in data analysis and are widely used in real-world projects. They allow us to process and analyze data efficiently.

```python
df = pd.read_csv(r"/content/Patient_Data.csv")

df.info()
df.head()
```

The dataset is loaded using the read_csv() function from pandas.

The info() function provides a summary of the dataset, including:
- Number of rows and columns
- Data types of each column
- Presence of missing values

The head() function displays the first few rows of the dataset, helping us understand how the data is structured.

Insight:
At this stage, we identify which columns are useful for analysis and check for missing or inconsistent data. This step is crucial because understanding the dataset before processing helps avoid errors in later stages.

```python
df = df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
df.head()
```

From the complete dataset, only relevant columns are selected:
PatientID, Department, Doctor, and BillAmount.

These columns are directly related to billing and analysis. Other columns such as administrative details are not required for this task.

Insight:
Reducing the dataset to only relevant columns improves performance and clarity. It also ensures that analysis is focused only on meaningful attributes, which is an important practice in data analysis workflows.

# Already handled by selecting relevant columns

Administrative columns like ReceptionistID and CheckInTime are excluded from the dataset.

These columns do not contribute to billing analysis and may introduce unnecessary complexity.

Removing irrelevant data helps in simplifying the dataset and improves efficiency during processing and analysis.

```python
dept_bill = df.groupby('Department')['BillAmount'].sum()
print(dept_bill)
```

The groupby() function is used to group data based on the Department column.

For each department, the total bill amount is calculated using the sum() function.

This operation aggregates the data and provides a summary of revenue generated by each department.

Insight:
This analysis helps identify which departments contribute the most to hospital revenue.
For example, departments like Cardiology or Neurology may show higher billing due to specialized treatments.

This information can be used by hospital management to:
- Allocate resources efficiently
- Identify high-performing departments
- Plan business strategies

```python
df.drop_duplicates(subset='PatientID', inplace=True)
```

Duplicate records are removed based on PatientID.

In real-world hospital data, a patient may appear multiple times due to follow-ups or repeated visits. However, for this analysis, we want to consider each patient only once.

Insight:
Removing duplicates ensures that data is not counted multiple times, which could lead to incorrect analysis such as inflated billing values or misleading trends.

This step is critical for maintaining data accuracy and reliability.

```python
## Handle Missing Values
mean_bill = df['BillAmount'].mean()
df['BillAmount'] = df['BillAmount'].fillna(mean_bill)
```

Missing values in the BillAmount column are replaced with the mean of the column.

The mean is calculated using the mean() function, and missing values are filled using fillna().

Insight:
Handling missing values is essential because they can cause errors in calculations and visualizations.

Using the mean ensures that the overall distribution of data remains balanced without removing rows.
This method is commonly used when the missing data is numerical and does not follow a specific pattern.

Proper handling of missing values improves data quality and ensures accurate analysis.

## Merge Billing Dataset

```python
billing = pd.read_csv(r"/content/Billing_Data.csv")

merged_df = pd.merge(df, billing, on='PatientID', how='inner')
merged_df.head()
```

The patient dataset is merged with the billing dataset using the PatientID column as a common key.

An inner join is used, which means only matching records from both datasets are included.

Insight:
Merging datasets is a fundamental operation in data analysis. It allows us to combine information from multiple sources into a single dataset.

In this case, merging provides a complete view of each patient, including both personal and billing details.

This combined dataset can be used for advanced analysis such as:
- Revenue analysis
- Doctor performance
- Patient spending patterns

```python
new_patients = pd.DataFrame({
    'PatientID': [201, 202],
    'Department': ['Orthopedics', 'Cardiology'],
    'Doctor': ['Dr. X', 'Dr. Y'],
    'BillAmount': [8000, 9000]
})

df = pd.concat([df, new_patients], axis=0)
df.tail()
```

New patient records are added to the dataset using row-wise concatenation.

The concat() function is used with axis=0, which appends rows to the existing dataset.

Insight:
This simulates real-world scenarios where new data is continuously generated and added to existing datasets.

Handling such dynamic data is an important skill in data analysis, especially in industries like healthcare, finance, and e-commerce.

```python
# ADD NEW COLUMNS
extra_columns = pd.DataFrame({
    'InsuranceCovered': ['Yes'] * len(df),
    'FinalAmount': df['BillAmount'] * 0.9
})

df = pd.concat([df, extra_columns], axis=1)
df.head()
```

New columns are added using column-wise concatenation.

The InsuranceCovered column indicates whether insurance is applied.
The FinalAmount column represents the payable amount after applying a 10% reduction.

Insight:
Adding new columns is a form of feature engineering, which is widely used in data analysis and machine learning.

The FinalAmount helps in understanding actual revenue after insurance deductions, which is more meaningful than raw billing data.

This step enhances the dataset and makes it more useful for decision-making.
